### Matplotlib 에서 한글 폰트 사용하기

```python
import matplotlib as mpl
import matplotlib.pyplot as plt 
import matplotlib.font_manager as fm
```

```python
# minus 표시
mpl.rcParams['axes.unicode_minus'] = False
```

```python
[(f.name, f.fname) for f in fm.fontManager.ttflist if 'Malgun' in f.name]
```

```python
plt.rcParams["font.family"] = 'Malgun Gothic'
plt.rcParams["font.size"] = 10
```

```python
%matplotlib inline
```

```python
import numpy as np 

xs, ysin, ycos = [], [], []
for x in range(360):
    xs.append(x)
    ysin.append(np.sin(np.pi * x / 180))
    ycos.append(np.cos(np.pi * x / 180))

xs = np.arange(0, 2*np.pi, 0.01)
ysin = np.sin(xs); ycos = np.cos(xs)
plt.plot(xs, ysin, label='사인 곡선')
plt.plot(xs, ycos, label='코사인 곡선')
plt.xlim([-0.1, 2*np.pi+0.1]); plt.ylim([-1.2, 1.2])
plt.xlabel('X-Value'); plt.ylabel('Y-Value')
plt.title('삼각함수')
plt.legend(loc='center') # label을 범례로 표시
plt.axhline(y=0, color='k', linewidth=0.5) # 수평선(x축)
plt.axvline(x=0, color='k', linewidth=0.5) # 수직선(y축)
plt.axvline(x=360, color='k', linewidth=0.5)
plt.show()
```
